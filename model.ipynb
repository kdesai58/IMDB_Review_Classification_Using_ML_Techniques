```python
lemmatizer=WordNetLemmatizer()
```

```python
stopwords.words('english')
```

```python
corpus = []
for i in range(0, len(X)):
    review = re.sub('[^a-zA-Z0-9]', ' ', X[i])
    review = review.lower()
    review = review.split()
    
    review = [lemmatizer.lemmatize(word) for word in review if not word in (stopwords.words('english')+['br'])]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus[0]
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
words=[]
for i in range(len(corpus)):
    sent_token=sent_tokenize(corpus[i])
    for sent in sent_token:
        a = simple_preprocess(sent)
        words.append(simple_preprocess(sent))   # simple_preprocess() will lower the token
```

```python
def avg_word2vec(para):
    return np.mean([model.wv[word] for word in para if word in model.wv.index_to_key],axis=0)
```

```python
Xn = np.empty((len(words),100))

for i in range(len(words)):
    a = avg_word2vec(words[i])
    # print(a)
    Xn[i] = a
```

```python
## train and test split

X_train, X_test, y_train, y_test = train_test_split(Xn, y, test_size = 0.25, random_state = 0)
```

```python
rf_classifier = RandomForestClassifier(random_state=0)
sgd_classifier = SGDClassifier(loss='log', random_state=0)
xgb_classifier = XGBClassifier(objective='binary:logistic', use_label_encoder=False, eval_metric='logloss', random_state=0)
```

```python
voting_classifier = VotingClassifier(estimators=[
    ('rf', rf_classifier),
    ('sgd', sgd_classifier),
    ('xgb', xgb_classifier)
], voting='soft')
```

```python
voting_classifier.fit(X_train,y_train)
y_pred = voting_classifier.predict(X_test)
```
